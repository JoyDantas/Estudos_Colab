```python
import pandas as pd
```

```python
tiktok_itaucard = pd.read_csv('/content/drive/MyDrive/historico - tiktok/TiktokAds_Ad Performance_DPZT _ Itaucard_20210101-20220101.csv', sep=";")
tiktok_institucional = pd.read_csv('/content/drive/MyDrive/historico - tiktok/TiktokAds_Ad Performance_DPZT _ ItauÃÅ Institucional_20210101-20220101.csv', sep=';')
tiktok_mc = pd.read_csv("/content/drive/MyDrive/historico - tiktok/TiktokAds_Ad Performance_DPZT _ McDonald's_20210101-20220101.csv", sep=";")
tiktok_natura = pd.read_csv('/content/drive/MyDrive/historico - tiktok/TiktokAds_Ad Performance_DPZT _ Natura_20210101-20220101.csv', sep=";")
```

```python
concat_frames = [tiktok_itaucard, tiktok_institucional, tiktok_mc, tiktok_natura]
```

```python
consolidado_tiktok = pd.concat(concat_frames)
```
