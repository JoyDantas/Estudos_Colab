#Podemos utilizar o **método map** para transformar dados usando um mapeamento definido por um dicionário

```python
df["year"] = df['data'].map(lambda x: str(str(x)[0:4])if x is not np.NaN else np.NaN) ## pegando o ano da data
df["dia"] = df['data'].map(lambda x: int(str(x)[-2:])if x is not np.NaN else np.NaN) ## pegando o dia da data
df["mes"] = df['data'].map(lambda x: int(str(x)[5:7])if x is not np.NaN else np.NaN) ## pegando o mês da data
```
